# Deep Learning for Hyperspectral Image Classification

### Imports

```python
%matplotlib inline

import numpy as np
import pandas as pd

import matplotlib.pyplot as plt

from scipy.io import loadmat
```

### Data Ingest

Hyperspectral data from: https://www.kaggle.com/datasets/syamkakarla/pavia-university-hsi

The data is in .mat format, a binary MATLAB format. We convert this to a Pandas dataframe for further processing.

```python
# from .mat format to numpy arrays
data_raw = loadmat('../data_hsi/PaviaU.mat')['paviaU']
data_class = loadmat('../data_hsi/PaviaU_gt.mat')['paviaU_gt']
```

```python
print(type(data_raw), data_raw.shape, type(data_raw), data_class.shape, (610 * 340))
```

```python
# from numpy arrays to flat pandas dataframe
df = pd.DataFrame(data_raw.reshape(data_raw.shape[0]*data_raw.shape[1], -1))
df.columns = [f'band{i}' for i in range(1, df.shape[-1]+1)]
df['class'] = data_class.ravel()

print(df.shape)
print(df.columns)
display(df.head(2))
display(df.tail(2))
```

### Inspect class label distribution

```python
df['class'].value_counts()
```

### Viz-1

```python
fig = plt.figure(figsize = (20, 12))

ax = fig.subplots(2, 5)

for i in range(2):
  for j in range(5):
    c = np.random.randint(103)
    ax[i][j].imshow(data_raw[:, :, c], cmap='nipy_spectral')
    ax[i][j].axis('off')
    ax[i][j].title.set_text(f"Band - {c}")
    c+=1

plt.tight_layout()
plt.show()

# see reference 2
```

### Ground Truth Viz-1

```python
def plot_data(data):
  fig = plt.figure(figsize=(12, 10))
  plt.imshow(data, cmap='nipy_spectral')
  plt.colorbar()
  plt.axis('off')
  plt.show()

# see reference 2
```

```python
plot_data(data_class)
```
